> 확률적 추세가 있다고 얘기하기엔 좀 애매함 / 있다고 하기에도 미심쩍음 -> 단위근 검정

> PACF의 경우 3차에서 절단 -> AR(3)?

`-` 모형 적합도 검정 : 포트맨토 검정

$H_0 : \rho_1 = \cdots = \rho_k = 0$ -> rate가 백색잡음과정인가?

```r
Box.test(dt$rate, lag = 1, type = "Ljung-Box")
Box.test(dt$rate, lag = 6, type = "Ljung-Box")
Box.test(dt$rate, lag = 12, type = "Ljung-Box")
```

> 셋 전부 귀무가설 기각 -> rate는 WN가 아님

`-` 단위근 검정 : 차분이 필요한가?

```r
z = rnorm(100) ## WN
cz = cumsum(z) ## random walk
```

```r
tsdisplay(z)
```

> 일단 백색잡음과정이라 할 수 있음.

```r
tsdisplay(cz)
```

> 뭔가 결정적 추세처럼 보이지만, ACF는 천천히 / PACF는 1시차만 유효함.
>
> ACF 그림만 보면, 단위근이 있는지 없는지 애매할 수 있음.

`type = c("nc", "c", "ct")`

* `nc` : 평균이 0
* `c` : 평균이 0이 아닌 것
* `ct` : 평균이 0이 아니고, 추세가 있음

*WN*

$H_0 : \phi_1$ -> 단위근이 있다.

```r
adfTest(z, lags = 1, type = "nc") ## 평균이 0을 기준으로 만들어졌으니까 일단은...
```

p-value가 유의수준 0.05보다 작으므로 귀무가설 기각 -> 단위근이 없다. -> 차분이 필요없다.

*Random Walk*

```r
adfTest(cz, lags = 1, type = "nc")
```

p-value가 유의수준 0.05보다 크므로 귀무가설 기각 못함 -> 단위근이 있음 -> 차분을 해야함

> 하지만, 똑같은 Random Walk임에도 불구하고 검정 결과가 다르다. 차분이 필요 없다고 판단할 수도 있다.

```r
## 단위근 검정 H0 : 단위근이 있다.
## 상수항, 추세 : (nc, c, ct)
adfTest(dt$rate, lags = 1, type = 'nc')
```

> 확률적 추세가 없음. 차분은 필요하지 않음.

> 거의 0에 가까운 평균

### 모형 적합

`order = c(p, d, q)`

`seasonal = list(order = c(P, D, Q), period = s)`

`include.mean = T` : 평균 포함 여부

`method = c("CSS-ML", "ML", "CSS")` : `CSS`(조건부 최소제곱법)로 초기값-`ML`(가능도비)로 추정

```r
fit1 = arima(dt$rate, order = c(3, 0, 0))
fit1
```

> intercept는 평균에 해당하는 것, drift.
>
> $\text{intercept} = \delta / (1-\phi_1-\phi_2-\phi_3)$

```r
-0.0606*(1-1.9691+1.3651-0.3394)
```

**MA term에서의 차이점**

$$Z_t = a_1 Z_{t-1} + \cdots + a_p Z_{t-p} + \epsilon_t + b_1 \epsilon_{t-1} + \cdots + b_q \epsilon_{t-q}$$

```r
summary(fit1)
```

AIC는 작을수록 좋고... ACF1 등등...

`-` 통계적 유의성 확인

```r
lmtest::coeftest(fit1)
```

> 각각의 추정량 / 유의성이 나와있음.
>
> intercept는 유의하지 않았으므로 평균은 제거를 해도 좋음.

`-` 평균을 빼고 적합

```r
fit2 = arima(dt$rate, order = c(3, 0, 0), include.mean = F)
summary(fit2)
```

> AIC가 작아졌으므로 평균을 빼는 게 더 좋음

```r
fit2$sigma2 ## 오차항의 분산 추정량
```

**잠정모형 : AR(3)**

$Z_t = 1.97 Z_{t-1} - 1.37 Z_{t-2} + 0.34 Z_{t-3} + \epsilon_t$

$\sigma^2 = 0.0353$

```r
resid = fit2$residuals
tsdisplay(resid)
```

> 튀어나온 애들이 좀 있긴 하지만... 어쨌든 그렇게까지 신경쓰일 정도로 튀어나온 것은 아니다. 문제가 없는?것 같다.

`-` 잔차의 정규성 검정

```r
tseries::jarque.bera.test(resid) ## H0 : normal distribution
```

```r
par(mfrow=c(1,2))
hist(resid)
qqnorm(resid, pch=16)
qqline(resid)
```

`-` 잔차의 포트맨토 검정 : $H_0 : \rho_1 = \cdots = \rho_k = 0$ -> 잔차가 백색잡음과정인가?

```r
portes::LjungBox(fit2, lags = c(6, 12, 18, 24))
```

### 과대적합 : AR(4), ARMA(3, 1)

```r
fit2 = arima(dt$rate, order = c(3,0,0)) ## AR(3)
fit3 = arima(dt$rate, order = c(4,0,0)) ## AR(4)
fit4 = arima(dt$rate, order = c(3,0,1)) ## ARMA(3, 1)
fit5 = arima(dt$rate, order = c(4,0,1)) ## ARMA(4, 1) -> 이러면 안돼요
```

```r
## 실제 coefficients가 유의한지 (phi / theta)
lmtest::coeftest(fit2)
lmtest::coeftest(fit3)
lmtest::coeftest(fit4)
lmtest::coeftest(fit5)
```

* ar(4)를 그렇게까지 할 필요는 없겠다~~
* arma(3,1)을 할 경우에는 나름대로 유의했다~~
* arma(4,1)은 하지마라, 공통근이 생겨서 모형이 이상해진다~~

```r
paste0("AIC for AR(3) = ", fit2$aic)
paste0("AIC for AR(4) = ", fit3$aic)
paste0("AIC for ARMA(3,1) = ", fit4$aic)
paste0("AIC for ARMA(4,1) = ", fit5$aic)
```

ARMA(3, 1)정도까진 할만한 것 같다! -> ARMA(3, 1)에 대해서 잔차 분석을 다시 해보면 된다!

`-` 과대적합된 모형 분석

```r
sarima(dt$rate, p=3, d=0, q=1)
```

> 달라지는 게 거의 없음. $\sigma^2$에 대한 추정량도 더 작아짐. 조금씩은 다 좋아지긴 했는데, 막 그렇게 많이 변한 게 아닌 것 같아서 굳이 이렇게 해야 하나라는 생각도 들긴 함

**선택에 따라서 ARMA(3, 1)을 선택할 수도 있겠으나, 차수가 커지는 것을 비선호하므로 그냥 AR(3)을 쓸 수도 있다. 교수님은 이쪽을 선호하심.**

> ACF는 천천히 감소 / PACF는 1시차에서만 유의하고 절단 형태 -> 차분이 필요하다.

`-` 단위근 검정

```r
## H0 : 단위근이 있다. (phi=1) H1 : |phi| < 1
adfTest(z, lags = 0, type = "c") ## 평균도 있음
adfTest(z, lags = 1, type = "c")
adfTest(z, lags = 2, type = "c")
```

> 단위근 검정 기각 못함 -> 차분이 필요함.
>
> 근데 `lags = 0, 1, 2`는 뭐여

```r
tseries::adf.test(z)
```

```r
diff_z = diff(z)
tsdisplay(diff_z)
```

> 0을 중심으로 이동, ACF/PACF가 모든 시차에서 유의하지 않음 -> WN가 됨.
>
> 그러니까 원자료는 Random Walk라는 거네?

```r
## 포투맨트검정 H0 : rho1=...=rho_k=0 -> WN임
Box.test(diff_z, lag=6, type = "Ljung-Box")
Box.test(diff_z, lag=12, type = "Ljung-Box")
```

```r
#평균
t.test(diff_z)
```

> WN라는 거 기각 불가

* $\nabla Z_t \sim WN(0, \sigma^2) \Rightarrow \nabla Z_t \sim ARIMA(0, 0, 0)$
* $Z_t \sim ARIMA(0, 1, 0)$

```r
## 모형적합 ARIMA(0,1,0)
fit <- arima(z,order=c(0,1,0))
summary(fit)
```

`-` 잔차검정

```r
resid = fit$residuals
tsdisplay(resid)
```

```r
## 정규성검정
tseries::jarque.bera.test(resid) ##JB test H0: normal
```

> 정규분포라기엔 좀 그럼/하지만 WN가 꼭 정규분포일 필요는 없음.

```r
par(mfrow=c(1,2))
hist(resid)
qqnorm(resid, pch=16)
qqline(resid)
```

```r
# 잔차의 포트맨토 검정 ## H0 : rho1=...=rho_k=0
portes::LjungBox(fit, lags=c(6,12,18,24)) ## 얘가 책에 있는 결과랑 같음. 어떤 걸 써도 상관 ㄴ
```

> 잔차의 WN 가정에 문제가 없음.

```r
## 잔차 검정
astsa::sarima(z, p=0, d=1, q=0)
```

> 대충 뭐 정규분포...
>
> 모든 시차에서 다 자기상관함수가 0임. WN임.

### 분해법을 사용하여 결정적 추세 제거

```r
t = 1:length(z)
fit1 = lm(z~t)
summary(fit1)
```

$\hat T_t = 186.35 + 4.07 t$

$Z_t = T_t + I_t = \beta_0 + \beta_1 t + \epsilon_t$

```r
et <- z - fitted(fit1) #추세 제거
forecast::tsdisplay(et, main = "추세를 제거한 여성근로자 수의 시계열 그림과 SACF/SPACF")
```

> ACF는 지수적으로 감소 / PACF는 1시자에서 절단 형태
>
> 오차항에 AR(1)을 적합할 수도... 단위근 검정을 할수도...

`-` 잔차 단위근 검정

```r
adfTest(et, lags = 0, type = "nc") ## 추세 제거했고, 평균이 0이니까
adfTest(et, lags = 1, type = "nc")
adfTest(et, lags = 2, type = "nc")
```

> 귀무가설 기각 -> 단위근 없음 -> 차분이 필요하지 않음.
>
> 기각을 해도 될 것 같기도 하고, 안해도 될 것 같기도 하고...

$\epsilon_t \sim AR(1)$

```r
fit_et = arima(et, order = c(1,0,0), include.mean = F)
summary(fit_et)
```

> $\phi$가 1에 가까운 값을 가지므로, ACF가 아주 천천히 감소. 하지만 절대값이 1보다 작으니까 정상시계열은 맞음

```r
lmtest::coeftest(fit_et)
```

> coefficient도 유의함...

`-` 결정적 추세를 분해법을 사용하여 제거한 후, 불규칙 성분에 대해 모형 적합한 결과

$Z_t = \beta_0 + \beta_1 \times t + \epsilon_t : \hat Z_t = 186.35 + 4.07 t$

$\epsilon_t = \phi \epsilon_{t-1} + u_t, ~ \{\epsilon_t\}\sim AR(1) ~ (\mu = 0, \hat\phi=0.91)$

$u_t \sim WN(0, \sigma^2)$

```r
#hat_zt = hat_Tt + hat_et
hat_zt = fitted(fit) + fitted(fit_et) ## 선형 추세 모형 + AR(1) 모형
ts.plot(z, hat_zt, col=1:2, lty=1:2, lwd=2)
legend("topleft", c("z","hat_z"), col=1:2, lty=1:2, lwd=2)
```

> 대충 이렇게 예측함...

`-` $u_t$가 WN인지 검정

```r
#### ut에 대한 잔차검정
ut <- resid(fit_et)
forecast::tsdisplay(ut)
```

```r
## 정규성검정
tseries::jarque.bera.test(ut) ##JB test H0: normal
```

```r
par(mfrow=c(1,2))
hist(ut)
qqnorm(ut, pch=16)
qqline(ut)
```

> 하나의 이상치 때문에 정규분포가 아니라고 나옴. (딱히 상관은 없음)

```r
# 잔차의 포트맨토 검정 ## H0 : rho1=...=rho_k=0
portes::LjungBox(fit_et, lags=c(6,12,18,24))
```

> 잔차는 백색잡음과정을 따른다는 귀무가설 기각 불가 -> WN임

```r
## 잔차 검정
astsa::sarima(et, p=1, d=0, q=0)
```

> ACF가 모든 차수에서 다 0이고, `Ljung-Box statistics`가 0.05보다 다 크다면 WN임
>
> 최종 모형이라고 주장이 가능함.

> $\text{intercept}$가 안나옴.

`-` 이렇게 하면 될것 같아요.

$Z_t = Z_{t-1} + \varepsilon_t$ : random_walk

$Z_t = \delta + Z_{t-1} + \varepsilon_t$ : random walk with drift <- 이걸 하고 싶은데, 자동으로 안해줌.

> $(1-B)Z_t = \delta + \varepsilon_t$ <- 백색잡음을 먹여야 함.

```r
fit2 = arima(diff_z, order = c(0,0,0))
summary(fit2)
```

$Z_t = \delta + Z_{t-1} + \varepsilon_t$ : random walk with drift

$\varepsilon_t \sim WN(0, \sigma^2)$

$\hat \delta = 4.215, ~ \hat \sigma^2 = 127.8$
