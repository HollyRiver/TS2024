## 2. 이동평균법

**왕의 평균 수명 데이터**

```r
kings=scan("http://robjhyndman.com/tsdldata/misc/kings.dat",skip=3)
```

```r
kings_ts = ts(kings)
plot(kings_ts, lwd = 2)
```

`-` `window = 3`

```r
kings_ts_SMA3 = SMA(kings_ts, n = 3) ## 굳이 time series를 넣어야 하는 것은 아님
plot.ts(kings_ts, lwd = 2)
lines(kings_ts_SMA3, col = 'red', lty = 2, lwd = 2)
legend('topleft', legend = c('original', expression(m==3)),
      col = c("black", "red"), lty = c(1, 2), lwd = 2)
```

```r
## vs windows = 10
kings_ts_SMA3 = SMA(kings_ts, n = 3)
kings_ts_SMA10 = SMA(kings_ts, n = 10)
plot.ts(kings_ts, lwd = 2)
lines(kings_ts_SMA3, col = 'red', lty = 2, lwd = 2)
lines(kings_ts_SMA10, col = 'skyblue', lty = 4, lwd = 2)
legend('topleft', legend = c('original', expression(m==3), expression(m==10)),
      col = c("black", "red", "skyblue"), lty = c(1, 2, 4), lwd = 2)
```

`m` 값이 달라짐에 따라 평활 정도가 달라진다 ㅇㅇ

### A. 모형 평가

예측을 하기 위한 평활은 기본적으로 아니긴 함...

```r
kings_ts_SMA3
```

```r
## m = 3
mean((kings[-1] - kings_ts_SMA3[-42])^2, na.rm = T) ## MSE : e_t(1) = Z_t+1 - hat Z_t, t = 3, 4, ...
mean(abs(kings[-1] - kings_ts_SMA3[-42]), na.rm = T) ## MAE
mean(abs(kings[-1] - kings_ts_SMA3[-42])/kings[-1], na.rm = T)*100 ## MAPE
```

> 4번째 관측치랑 첫번째 예측값을 비교해야 함. t = 4일 때부터 예측 시작

```r
## m = 10
mean((kings[-1] - kings_ts_SMA10[-42])^2, na.rm = T) ## MSE : e_t(1) = Z_t+1 - hat Z_t, t = 10, 11, ...
mean(abs(kings[-1] - kings_ts_SMA10[-42]), na.rm = T) ## MAE
mean(abs(kings[-1] - kings_ts_SMA10[-42])/kings[-1], na.rm = T)*100 ## MAPE
```

**예측의 관점에서 윈도우 값을 조정하면서 예측오차를 가장 작게 하는 모형을 선택해야 한다.**

### B. mindex data

```r
z = scan('mindex.txt')
mindex = ts(z, start = c(1986, 1), frequency = 12)
mindex
```

```r
mindex_sma3 = SMA(mindex, n = 3)
mindex_sma10 = SMA(mindex, n = 10)
```

```r
plot.ts(mindex, lwd = 2)
lines(mindex_sma3, col = 'red', lty = 2)
lines(mindex_sma10, col = 'skyblue', lty = 2)
```

```r
z = scan('depart.txt')
mindex = ts(z, start = c(1986, 1), frequency = 12)
mindex
```

```r
plot.ts(mindex, lwd = 2)
mindex_sma3 = SMA(mindex, n = 3)
mindex_sma12 = SMA(mindex, n = 12) ## 반복 주기가 12개월이므로, n = 12로 하면 plat한 평활 추세를 가져올 수 있다.
lines(mindex_sma3, col = 'red', lty = 2)
lines(mindex_sma12, col = 'skyblue', lty = 2, lwd = 2)

legend("topright", legend = c("original", expression(m==3), expression(m==12)),
      col = c('black', 'red', 'skyblue'), lty = c(1, 2, 2))
```

> **계절성분만 추출할 때, Simple Moving Average를 사용할 수 있다!!!**
