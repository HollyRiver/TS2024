### A. 모형 평가

예측을 하기 위한 평활은 기본적으로 아니긴 함...

```r
kings_ts_SMA3
```

```r
## m = 3
mean((kings[-1] - kings_ts_SMA3[-42])^2, na.rm = T) ## MSE : e_t(1) = Z_t+1 - hat Z_t, t = 3, 4, ...
mean(abs(kings[-1] - kings_ts_SMA3[-42]), na.rm = T) ## MAE
mean(abs(kings[-1] - kings_ts_SMA3[-42])/kings[-1], na.rm = T)*100 ## MAPE
```

> 4번째 관측치랑 첫번째 예측값을 비교해야 함. t = 4일 때부터 예측 시작

```r
## m = 10
mean((kings[-1] - kings_ts_SMA10[-42])^2, na.rm = T) ## MSE : e_t(1) = Z_t+1 - hat Z_t, t = 10, 11, ...
mean(abs(kings[-1] - kings_ts_SMA10[-42]), na.rm = T) ## MAE
mean(abs(kings[-1] - kings_ts_SMA10[-42])/kings[-1], na.rm = T)*100 ## MAPE
```

**예측의 관점에서 윈도우 값을 조정하면서 예측오차를 가장 작게 하는 모형을 선택해야 한다.**

## 3. 단순지수평활법

```r
z = scan("mindex.txt")
mindex = ts(z, start = c(1986, 1), frequency = 12)
```

```r
tmp_dat = data.frame(day = seq.Date(as.Date("1986-01-01"), by = 'month', length.out = length(z)), ind = z)
head(tmp_dat)
```

```r
ggplot(tmp_dat, aes(day, ind)) +
    geom_line(col = 'skyblue', lwd = 2) +
    geom_point(col = 'steelblue', cex = 2) +
    ggtitle("중간재 출하지수") +
    theme_bw() +
    theme(plot.title = element_text(size = 30),
         axis.title = element_blank())
```

`HoltWinters()` : 모든 지수 평활이 가능

> `HoltWinters()` : `x = data, alpha = NULL, beta = NULL, gamma = NULL,`
>
> level only model에서는 첫번째 값을 초기값으로, 트랜드는 두번째 값으로, 기울기가 있으면 두번째 값에서 첫번째 값을 뺀 것을 초기값으로, ...

`-` 단순지수평활 $\alpha = 0.3$

```r
fit0 = HoltWinters(mindex,
                   alpha = 0.3, beta = FALSE, gamma = FALSE) ## default : NULL, 직접 지정해줘야 배제됨.
fit0
```

> `Coefficients` : 마지막 시점의 평활값, 마지막 평활값으로 모든 예측을 진행

```r
ls(fit0)
```

```r
head(fit0$fitted)
```

> 단순지수평활이고, 이는 level밖에 없는 모형이므로 level이 곧 평활값이 된다.

```r
head(cbind(fit0$fitted, mindex))
```

* 초기값은 첫 번째 관측값 `9.3`
* 두번째 값부터 아래 수식을 사용

$$\begin{align} S_1^{(1)} & = \omega \times Z_1 + (1-\omega)S_0^{(1)} = Z_1 \because S_0^{(1)} = Z_1 \\
S_2^{(1)} & = \omega \times Z_2 + (1-\omega)S_0^{(1)} \\
S_3^{(1)} & = \omega \times Z_3 + \omega (1-\omega) \times Z_2 + (1-\omega)^2S_0^{(1)}\end{align}$$

```r
plot(fit0, lwd = 3, cex.main = 2)
legend('topright', legend = c("observed", "ses"), lty = 1, lwd = c(1, 3), col = 1:2)
```

```r
tail(cbind(fit0$fitted, mindex))
```

* 예측 갱신

$$S_{100}^{(1)} = \omega \times Z_{100} + (1-\omega) \times S_{99}^{(1)}$$

```r
0.3*11.4 + 0.7*11.148
```

`ses()` : 단순지수평활만 가능, 예측까지 수행할 수 있음ㅁ

```r
fit01 = ses(mindex, alpha = 0.3, initial = 'simple', h = 10) ## h : Number of forecastings
```

```r
fit01
```

```r
summary(fit01)
```

```r
plot(fit01)
```

> 모델을 바로 그래프에 넣으면 예측만 수행한다. 신뢰구간을 표시한다.

### 여러 평활상수에 따른 변화

```r
w = c(seq(0.1, 0.8, 0.1), seq(0.81, 0.99, 0.01))
```

```r
SSE_ses = sapply(w, function(alpha) HoltWinters(mindex, alpha = alpha, beta = FALSE, gamma = FALSE)$SSE)

plot(w, SSE_ses, type = 'o', xlab = "weight", ylab = "SSE", pch = 16)
plot(w[-(1:7)], SSE_ses[-(1:7)], type = 'o', xlab = "weight", ylab = "SSE", pch = 16)
```

* SSE가 가장 작은 w를 반영하여 모델 생성

```r
fit1 = ses(mindex, alpha = w[which.min(SSE_ses)], initial = 'simple', h = 6)
```

```r
plot(fit1)
lines(fitted(fit1), col = 'red', lty = 2)
```

> 그러면 `alpha` 최적값을 구할 때, 항상 수동으로 해야 하나??? 당연히 모듈화되어 있음.

```r
fit0_w = HoltWinters(mindex, beta = FALSE, gamma = FALSE)
fit0_w
```

> 파라미터가 NULL이면 1시차 후 오차에 대한 제곱합(SSE)을 줄이는 방향으로 알아서 최적화하여 값을 산출한다.

### 잔차

```r
plot(resid(fit0_w))
abline(h = 0)
```

`-` 오차의 평균이 0인가? $H_0 : \mu = 0 ~ vs. ~ H_1 : \mu \neq 0$

```r
t.test(resid(fit0_w))
```

`-` 오차가 독립인가?

```r
dwtest(lm(resid(fit0_w)~1), alternative = 'two.sided') ## 오차에 대해서 진단하고자 하면 ~1으로 회귀모형을 잡아주면 된다(절편만 있는 회귀모형)
```

```r
## 0.3을 쓰는 건 어떨까?
fit2 = ses(mindex, alpha = 0.3, h = 6)
plot(fit2)
lines(fitted(fit2), col = 'red', lty = 2)
```

```r
plot(resid(fit2))
abline(h = 0, lty = 2)
```

## 4. 이중지수평활법

* 계절성분은 없지만 추세가 있는 경우에 사용하는 평활법

```r
data(valueofstocks)
```

```r
stock = valueofstocks[, 1]
```

```r
plot(stock)
```

### A. 1모수 이중지수평활($\alpha = \beta$)

`holt` : 이중지수평활 버전 `ses`

```r
fit4 = holt(stock, alpha = 0.6, beta = 0.6, h = 6)
fit4$model
```

```r
plot(fit4, ylab = "", xlab = "", lty = 1)
lines(fitted(fit4), col = 'red', lty = 2)
```

```r
fit4_tmp = holt(stock, alpha = 0.2, beta = 0.2, h = 6)
plot(fit4_tmp, ylab = "", xlab = "", lty = 1)
lines(fitted(fit4_tmp), col = 'red', lty = 2)
```

> `alpha`값을 더 높여서 이전 데이터에 더 가중치를 둘 수 있음 ㅇㅇ

### B. 2모수 이중지수평활

```r
fit5 = holt(stock, h = 6)
fit5$model
```

```r
fit50 = HoltWinters(stock, alpha = 0.2, beta = 0.2, gamma = FALSE)
fit50
```

```r
fit4_tmp$model
```

> `HoltWinters`는 SSE로 최적화하는데, `Holt`는 AIC로 최적화해서 계수값이 좀 다름...

* `HoltWinters`는 예측을 못하나요?

```r
fit50 = HoltWinters(stock, gamma = FALSE)
predict(fit50, n.ahead = 12, prediction.interval = TRUE, level = 0.95)
```

> 회귀분석에서 prediction하듯 하면 된다.

```r
plot(fit5, ylab = 'Index', xlab = 'year', main = expression("중간재 출하지수와 이중지수평활값 : "~alpha~","~beta~" : estimated"))
lines(fitted(fit5), col = 'red', lty = 2)
```

```r
plot(resid(fit5))
abline(h = 0, lty =2)
```

> 이분산성도 있어보이고, 독립성도 글쎄?

## 5. 계절지수평활

```r
z = scan("koreapass.txt")
pass = ts(z, start = c(1981, 1), frequency = 12)
```

```r
plot.ts(pass)
```

> 추세에 이어 계절속성까지 평활을 해야 할 것으로 보임 ㅇㅇ

`-` 가법모형

```r
fit_hw = HoltWinters(pass, seasonal = "additive") ## default
## seasonal model의 경우, gamma = TRUE인 경우 인풋 데이터가 무조건 time series여야 한다.
fit_hw
```

> a는 level, b는 추세, 나머지는 계절 성분.
>
> `a` : $\hat T_n$, `b` : $\hat \beta_{1, n}$

```r
plot(fit_hw)
```

```r
predict(fit_hw, n.ahead = 12, prediction.interval = T, level = 0.95)
```

* **편리한 forecasting을 위한 라이브러리 사용**

```r
## forecast library, HoltWinters는 기본 함수임
fit6 = hw(pass,
          alpha = fit_hw$alpha, beta = fit_hw$beta, gamma = fit_hw$gamma,
          seasonal = "additive", initial = 'simple', h = 12) ## initial = simple : 최적화 안하고 한 값으로 고정
fit6$model
```

```r
plot(fit6)
lines(fit6$fitted, lty = 2, col = 'red')
```

> 예측 구역을 편하게 나타나도록 한다.

`-` 승법모형

```r
fit_hw_m = HoltWinters(pass, seasonal = "multiplicative")
fit_hw_m
```

```r
fit_hw_m$fitted |> head()
```

> (level + trend) * season

```r
predict(fit_hw_m, n.ahead = 12, prediction.interval = TRUE, level = 0.95)
```

```r
fit7 = hw(pass,
          alpha = fit_hw_m$alpha,
          beta = fit_hw_m$beta,
          gamma = fit_hw_m$gamma,
          seasonal = 'multiplicative',
          initial = 'simple',
          h = 12)
fit7$model
```

```r
plot(fit7)
lines(fit7$fitted, lty = 2, col = 'red')
```
