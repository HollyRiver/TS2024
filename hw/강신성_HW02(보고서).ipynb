$$\begin{align} \notag n & = 15, \bar Z = 6 \\
\notag \hat{\gamma_0} & = \frac{1}{15} \underset{t=1}{\overset{15}{\sum}}(Z_t - 6)^2 = \frac{32}{15} = 2.1333 \\
\notag \hat{\gamma_1} & = \frac{1}{15} \underset{t=1}{\overset{14}{\sum}}(Z_t - 6)(Z_{t+1} - 6) = \frac{3}{15} = 0.2 \\
\notag \hat{\gamma_2} & = \frac{1}{15} \underset{t=1}{\overset{13}{\sum}}(Z_t - 6)(Z_{t+2} - 6) = \frac{-9}{15} = -0.6 \\
\notag \hat{\gamma_3} & = \frac{1}{15} \underset{t=1}{\overset{12}{\sum}}(Z_t - 6)(Z_{t+3} - 6) = \frac{-4}{15} = -0.2667
\end{align}$$

* $\text{SACF}, ~ \hat \rho_h ~ (h = 1, 2, 3)$

$$\begin{align} \notag \hat \rho_1 & = \frac{\hat\gamma_1}{\hat\gamma_0} = \frac{0.2}{32/15} = 0.09375 \\
\notag \hat \rho_2 & = \frac{\hat\gamma_2}{\hat\gamma_0} = \frac{-0.6}{32/15} = -0.28125 \\
\notag \hat \rho_3 & = \frac{\hat\gamma_3}{\hat\gamma_0} = \frac{-4/15}{32/15} =-0.125 \\
\end{align}$$

* $\text{SPACF}, ~ \hat \phi_{kk} ~ (k = 1, 2)$

$$\begin{align} \notag \hat \phi_{11} & = \hat \rho_1 = 0.09375 \\

* 가설 $H_0 : \rho_h = 0 ~ vs. ~ H_1 : \rho_h \neq 0$
* 기각치 : $\frac{2}{\sqrt{15}} = 0.5164$
* 기각치에 대하여 검정통계량 $\hat\rho_1, \hat\rho_2, \hat\rho_3$ 모두 기각치보다 낮으므로, 귀무가설을 기각할 수 없었다. 따라서 1, 2, 3차 자기상관관계는 존재하지 않는다.

### 4. 다음의 모형들에 의해 설명되는 확률과정 $\{Z_t\}$는 정상성을 갖는가? (단, $\epsilon_t \sim WN(0, 1)$)

---

#### (1) $Z_t = \epsilon_t - \epsilon_{t-1} - \epsilon_{t-2}$

$$\begin{align} \notag \mathbb{E}(Z_t) & = \mathbb{E}(\epsilon_t - \epsilon_{t-1} - \epsilon_{t-2}) = 0 \\ \notag \\
\notag Var(Z_t) & = Var(\epsilon_t - \epsilon_{t-1} - \epsilon_{t-2}) \\
\notag & = Var(\epsilon_t) + Var(\epsilon_{t-1}) + Var(\epsilon_{t-2}) \\
\notag & = 3 \\ \notag \\
\notag Cov(Z_t, Z_{t+1}) & = Cov(\epsilon_t - \epsilon_{t-1} - \epsilon_{t-2}, ~ \epsilon_{t+1} - \epsilon_{t} - \epsilon_{t-1}) \\
\notag & = -Var(\epsilon_t) + Var(\epsilon_{t-1}) \\
\notag & = 0 \\
\notag Cov(Z_t, Z_{t+2}) & = Cov(\epsilon_t - \epsilon_{t-1} - \epsilon_{t-2}, ~ \epsilon_{t+2} - \epsilon_{t+1} - \epsilon_{t}) \\
\notag & = -Var(\epsilon_t) \\
\notag & = -1 \\
\notag Cov(Z_t, Z_{t+k}) & = 0 ~ (k > 2)\\
\notag \therefore Cov(Z_t, Z_{t+h}) & = \left\{ \begin{array}{lcl} -1, ~ h = 2 \\ 0, ~~~~ h = 1, 3, 4, 5, \cdots \end{array}\right.
\end{align}$$

&nbsp; 즉, 평균과 분산, 자기공분산이 시점에 따라 변화하지 않으므로, 확률과정 $\{Z_t\}$는 정상성을 가진다.

#### (2) $Z_t = \epsilon_t\epsilon_{t-1} + \epsilon_{t-2}$

$$\begin{align} \notag \mathbb{E}(Z_t) & = \mathbb{E}(\epsilon_t \epsilon_{t-1} + \epsilon_{t-2}) = Cov(\epsilon_t, ~ \epsilon_{t-1}) = 0 \\ \notag \\
\notag Var(Z_t) & = Var(\epsilon_t\epsilon_{t-1} + \epsilon_{t-2}) \\
\notag & = \mathbb{E}\big[\big(\epsilon_t\epsilon_{t-1} - \mathbb{E}(\epsilon_t\epsilon_{t-1})\big)^2\big] + Var(\epsilon_{t-2}) \\
\notag & = \mathbb E (\epsilon_t^2 \epsilon_{t-1}^2) + 1 \\
\notag & = Var(\epsilon_t) \times Var(\epsilon_{t-1}) + 1 \\
\notag & = 2 \\ \notag \\
\notag Cov(Z_t, Z_{t+1}) & = Cov(\epsilon_t\epsilon_{t-1}+\epsilon_{t-2}, ~ \epsilon_{t+1}\epsilon_{t}+\epsilon_{t-1}) \\
\notag & = Cov(\epsilon_t\epsilon_{t-1}, ~ \epsilon_{t+1}\epsilon_t) + Cov(\epsilon_t\epsilon_{t-1}, ~ \epsilon_{t-1}) \\
\notag & = \mathbb E\big[\big(\epsilon_t\epsilon_{t-1} - \mathbb E(\epsilon_t\epsilon_{t-1})\big)\big(\epsilon_{t+1}\epsilon_t - \mathbb E(\epsilon_{t+1}\epsilon_t)\big)\big] \\
\notag & ~~~~ + \mathbb E\big[\big(\epsilon_t\epsilon_{t-1} - \mathbb E(\epsilon_{t+1}\epsilon_t)\big)\big(\epsilon_{t-1} - \mathbb E (\epsilon_{t-1})\big)\big] \\
\notag & = \mathbb E (\epsilon_{t+1}\epsilon_t^2\epsilon_{t-1}) + \mathbb E(\epsilon_t \epsilon_{t-1}) \\
\notag & = 0 \\ \end{align}$$

$$\begin{align}
\notag Cov(Z_t, Z_{t+2}) & = Cov(\epsilon_t\epsilon_{t-1}+\epsilon_{t-2}, ~ \epsilon_{t+2}\epsilon_{t+1}+\epsilon_{t}) \\
\notag & = Cov(\epsilon_t\epsilon_{t-1}, ~ \epsilon_t) \\
\notag & = \mathbb E\big[\big(\epsilon_t\epsilon_{t-1} - \mathbb E(\epsilon_t\epsilon_{t-1})\big)\big(\epsilon_t - \mathbb E (\epsilon_t)\big)\big] \\
\notag & = \mathbb E(\epsilon_t^2 \epsilon_{t-1}) \\
\notag & = 0 \\
\notag & ~~~~~~ \vdots \\
\notag Cov(Z_t, Z_{t+h}) & = 0
\end{align}$$

&nbsp; 즉, 평균과 분산, 자기공분산이 시점에 따라 변화하지 않으므로, 확률과정 $\{Z_t\}$는 정상성을 갖는다.

#### (3) $Z_t = A \sin\Big(\frac23 \pi t + U\Big)$, 단, $A$는 평균이 $0$이고 분산이 $1$인 확률변수이고, $U$는 상수이다.

$$\begin{align} \notag \mathbb{E}(Z_t) & = \mathbb E [A\sin\Big(\frac23 \pi t + U\Big)] \\
\notag & = \sin\Big(\frac23 \pi t + U\Big) \mathbb E (A) \\
\notag & = 0 \\ \notag \\
\notag Var(Z_t) & = Var[A\sin\Big(\frac23 \pi t + U\Big)] \\
\notag & = \sin^2\Big(\frac23 \pi t + U\Big) Var(A) \\
\notag & = \sin^2\Big(\frac23 \pi t + U\Big) \\ \notag \\
\notag Cov(Z_t, Z_{t+h}) & = Cov(A\sin\Big(\frac23 \pi t + U\Big), ~ A\sin\Big(\frac23 \pi (t+h) + U\Big)) \\
\notag & = \sin\Big(\frac23 \pi t + U\Big)\times \sin\Big(\frac23 \pi (t+h) + U\Big) Var(A) \\
\notag & = \sin\Big(\frac23 \pi t + U\Big)\times \sin\Big(\frac23 \pi (t+h) + U\Big)
\end{align}$$

&nbsp; 즉, 분산, 자기공분산이 시점에 따라 변화하기 때문에, 확률과정 $\{Z_t\}$는 비정상 확률과정이다.

#### (4) $Z_t = A \sin (\pi t + U)$, 단 $A$는 평균이 $0$이고 분산이 $1$인 확률변수이고, $U$는 $[-\pi, \pi]$에서의 균등분포를 따르는 확률변수이다. 또한 $A, U$는 서로 독립이다.다.

* 먼저, 확률변수 $U$의 범위는 삼각함수의 한 주기에 해당하므로, $\mathbb E(\sin(\pi t + U)) = \mathbb E(\sin(U)) = \mathbb E(\cos(U)) = 0, ~ \mathbb E(\sin^2(\pi t + U)) = \frac12$이다.

$$\begin{align} \notag \mathbb E(Z_t) & = \mathbb E(A\sin(\pi t + U)) \\
\notag & = \mathbb E(A) \times \mathbb E(\sin(\pi t + U) \\
\notag & = 0 \\ \notag \\
\notag Var(Z_t) & = Var(A \sin(\pi t + U)) \\
\notag & = \mathbb E[A^2 \sin^2 (\pi t + U)] \\
\notag & = Var(A)\times \mathbb E(\sin^2(\pi t + U)) \\
\notag & = 1\times \frac12 = \frac12 \\ \notag \\
\notag Cov(Z_t, Z_{t+h}) & = Cov(A\sin(\pi t + U), ~ A\sin(\pi (t+h) + U)) \\
\notag & = \mathbb E\big(A^2\sin(\pi t + U) \times \sin(\pi (t +h) + U)\big) \\
\notag & ~~~~ - \mathbb E(A\sin(\pi t + U)) \times \mathbb E (A\sin(\pi (t+h) + U)) \\
\notag & = Var(A)\times \mathbb E\big(\sin(\pi t + U) \times \sin(\pi (t +h) + U)\big) \\
\notag & = \mathbb E \big[-\frac12\big\{\cos(\pi(2t + h) + 2U) - \cos(\pi h)\big\}\big] \\
\notag & = \frac12 \mathbb E (\cos(\pi h)) - \frac12 \mathbb E (\cos(\pi(2t + h) + 2U)) \\
\notag & = \frac12 \cos(\pi h)
\end{align}$$

&nbsp; 즉, 평균, 분산, 자기공분산이 시점에 따라 변화하지 않고, 자기공분산은 시차에만 의존하기 때문에, 확률과정 $\{Z_t\}$는 정상성을 가진다.

#### (5) $\left\{ \begin{array}{lcl} Z_t = \epsilon_t, ~~~~~~~~~ t : \text{짝수} \\ Z_t = \epsilon_t + 1, ~~ t : \text{홀수} \end{array}\right.$

$$\begin{align} \notag \mathbb E(Z_t) & = \left\{ \begin{array}{lcl} 0, ~~~~ t : \text{짝수} \\ 1, ~~~~ t : \text{홀수} \end{array}\right. \\ \notag \\
\notag Var(Z_t) & = Var(\epsilon_t + 1) = Var(\epsilon_t) = 1 \\ \notag \\
\notag Cov(Z_t, Z_{t+h}) & = Cov(\epsilon_t, \epsilon_{t+h}) ~~~~~~~~~~~~~~~~ (t, h : \text{짝수}) \\
\notag & = Cov(\epsilon_t, \epsilon_{t+h} + 1) ~~~~~~~~~ (t : \text{짝수}, h : \text{홀수}) \\
\notag & = Cov(\epsilon_t + 1, \epsilon_{t+h}) ~~~~~~~~~ (t : \text{홀수}, h : \text{홀수}) \\
\notag & = Cov(\epsilon_t + 1, \epsilon_{t+h} + 1) ~~ (t : \text{홀수}, h : \text{짝수}) \\
\notag & = 0
\end{align}$$

&nbsp; 시점의 홀ㆍ짝수에 따라 평균이 변하기 때문에, 확률과정 $\{Z_t\}$는 비정상 시계열이다.

#### (6) $Z_t = 1.3 Z_{t-1} + \epsilon_t$

$$\begin{align} \notag Z_t & = 1.3 Z_{t-1} + \epsilon_t \\
\notag & = 1.3(1.3 Z_{t-2} + \epsilon_{t-1}) + \epsilon_t \\
\notag & ~~~~~~~ \vdots \\
\notag & = \underset{i=0}{\overset{t-1}{\sum}}1.3^i \epsilon_{t-i} ~~~~~~~~ (Z_0 = 0) \\ \notag \\
\notag \mathbb E(Z_t) & = \mathbb E\bigg(\underset{i=0}{\overset{t-1}{\sum}}1.3^i \epsilon_{t-i}\bigg) = \underset{i=0}{\overset{t-1}{\sum}}1.3^i \mathbb E(\epsilon_{t-i}) \\
\notag & = 0 \\ \notag \\
\notag Var(Z_t) & = Var\bigg(\underset{i=0}{\overset{t-1}{\sum}}1.3^i \epsilon_{t-i}\bigg) = \underset{i=0}{\overset{t-1}{\sum}}1.3^i Var(\epsilon_{t-i}) \\
\notag & = \underset{i=0}{\overset{t-1}{\sum}}1.3^i \\ \notag \\
\notag Cov(Z_t, Z_{t+h}) & = Cov(\epsilon_t + 1.3\epsilon_{t-1} + \cdots + 1.3^{t-1} \epsilon_1 \\
\notag & ~~~~~~~~~~~~~~, \epsilon_{t+h} + 1.3 \epsilon_{t+h-1} + \cdots + 1.3^h\epsilon_t + \cdots + 1.3^{t+h-1} \epsilon_1) \\
\notag & = \underset{i=0}{\overset{t-1}{\sum}}1.3^{h+i} Var(\epsilon_{t-i}) \\
\notag & = \underset{i=0}{\overset{t-1}{\sum}}1.3^{h+i}
\end{align}$$

&nbsp; 분산과 자기공분산이 시간이 지남에 따라 증가하기 때문에, 확률과정 $\{Z_t\}$는 비정상 시계열이다.

$$\begin{align} \notag \mathbb E(Z_t) & = \frac12 \times 1 + \frac12 \times (-1) = 0 ~~~~~~~~~~~~ (t : \text{짝수}) \\
\notag \mathbb E(Z_t) & = \mathbb E(Z_{t-1}) = 0 ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~ (t : \text{홀수}) \\ \notag \\
\notag Var(Z_t) & = \mathbb E\big[(Z_t - \mathbb E(Z_t))^2\big] \\
\notag & = \mathbb E(Z_t^2) \\
\notag & = \frac12 \times 1^2 + \frac12 \times (-1)^2 \\
\notag & = 1 \\
\notag Var(Z_t) & = Var(Z_{t-1}) = 1 \\ \notag \\
\notag Cov(Z_t, Z_{t+1}) & = Cov(Z_t, Z_t) = Var(Z_t) = 1 ~~~~~~ (t : \text{짝수}) \\
\notag Cov(Z_t, Z_{t+1}) & = Cov(Z_{t-1}, Z_{t+1}) = 0 ~~~~~~~~~~~~~~~~~~~ (t : \text{홀수}) \\ \notag \\
\notag Cov(Z_t, Z_{t+2}) & = 0 ~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~ (t : \text{짝수}) \\
\notag Cov(Z_t, Z_{t+2}) & = Cov(Z_{t-1}, Z_{t+1}) = 0 ~~~~~~~~~~~~~~~~~~~ (t : \text{홀수}) \\
\notag \vdots \\
\notag Cov(Z_t, Z_{t+h}) & = 0
\end{align}$$
