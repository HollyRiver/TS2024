> &nbsp; 먼저, 전문기술행정직에 종사하는 여성근로자의 수는 지속적으로 증가하는 추세가 있는 것으로 파악된다. 하지만, 연 단위로 계절에 따른 반복적인 패턴이 존재한다고 보기는 어려울 것 같다. 따라서, 해당 자료에는 추세성분과 불규칙성분만이 있다고 판단하여 이중지수평활을 적용하는 것이 바람직하다고 생각한다.

`-` 평활법 적용 및 예측오차 분석

#### (2) 월별 수출액(단위:억$)
