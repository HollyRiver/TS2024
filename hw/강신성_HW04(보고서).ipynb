#### (5) 각 확률과정에 대해서 $\text{PACF} ~ \phi_{kk}, ~ k = 1, 2, 3, 4$를 계산하여라.

$$\begin{align} \notag \phi_{11} & = \rho_1 \\ \notag \\
\notag \phi_{22} & = \frac{\rho_2 - \rho_1^2}{1-\rho_1^2}
\end{align}$$

$$\begin{align}\notag \left\{ \begin{array}{lcl} \rho_1 = \phi_{k1} \rho_0 + \phi_{k2} \rho_1 + \cdots + \phi_{kk} \rho_{k-1} \\ \rho_2 = \phi_{41} \rho_1 + \phi_{42} \rho_0 \cdots + \phi_{kk} \rho_{k-2} \\ ~~~~~~~~~~~~~~~~~~~~~~~~~~~\vdots \\ \rho_k = \phi_{k1} \rho_{k-1} + \phi_{k2} \rho_{k-2} + \cdots + \phi_{kk} \rho_0 \end{array}\right.
\end{align}$$

> $Z_t \sim AR(1) ~ \text{process}$이므로, $\text{PACF}$는 1시차 이후로는 0이 된다.

> $\phi_{11} = \rho_1 = 0.8, ~ \phi_{22} = \cdots = \phi_{55} = 0$

### 3. $\epsilon_t$가 $WN(0, \sigma^2)$를 따를 때, 다음과 같은 확률과정 모형에 대하여 각 물음에 답하고, 답에 대한 이유를 밝혀라.

$$(1-B)^2(1-B^{12})Z_t = (1-0.5B)\epsilon_t$$

---

#### (1) $\{Z_t\}$는 정상시계열인가?

`No`

&nbsp; $\Phi(B) = (1-B)^2(1-B^{12})$의 모든 실근이 $-1, 1$로, 절대값이 1 이하이기 때문에 주어진 확률과정 모형은 비정상 시계열이다.

&nbsp; 또한, 해당 확률과정 모형은 2차 확률적 추세와 주기가 12인 확률적 계절 추세를 지니는 모형이므로, 계절 차분과 이차 차분을 전부 진행하지 않는다면 정상시계열이 될 수 없을 것이다.

#### (2) $W_t = (1-B)Z_t$는 정상시계열인가?

`No`

&nbsp; $W_t$에는 아직 1차 확률적 추세와 주기가 12인 확률적 계절 추세가 남아있으므로, 정상시계열이 아니다.

#### (3) $W_t = Z_t - 2Z_{t-1} + Z_{t-2}$는 정상시계열인가?

`No`

&nbsp; $W_t$에는 확률적 다항 추세가 완전히 제거되었지만, 아직 확률적 계절 추세가 남아있으므로 정상시계열이 아니다.

#### (4) $W_t = Z_t - Z_{t-12}$는 정상시계열인가?

`No`

&nbsp; $W_t$에는 확률적 계절 추세가 완전히 제거되었지만, 아직 2차 확률적 추세가 남아있기 때문에 정상시계열이 아니다.

#### (5) $W_t = (1-B)^2(1-B^{12})Z_t$는 정상시계열인가?

`Yes`

&nbsp; $W_t = (1-0.5B)\epsilon_t$는 $MA(1)$ 과정이므로, 정상성이 보장된다. 따라서 정상시계열이다.
